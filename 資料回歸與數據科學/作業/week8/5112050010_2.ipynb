# Decision Tree Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
m = 200
x = np.random.rand(m, 1)
y = 4 * (x - 0.5) ** 2 + np.random.randn(m, 1)/10
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.show()
```

## DecisionTreeRegressor()

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=1)
tree_reg.fit(x, y)
```

```python
tree_reg.score(x, y)
```

```python
from sklearn import tree

plt.figure(figsize=(5, 3))
tree.plot_tree(tree_reg, feature_names=['x1'], filled=True, rounded=True)
# plt.savefig('tree_regression_plot.pdf') 
plt.show()
```

```python
tree_reg.predict([[0.4]])
```

```python
tree_reg.feature_importances_
```

```python
def plot_regression_predictions(tree_reg, x, y, axes=[0, 1, -0.2, 1]):
  x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
  y_pred = tree_reg.predict(x1)
  
  plt.plot(x, y, "b.")
  plt.plot(x1, y_pred, "r-", linewidth=2, label=r"$\hat{y}$")

  plt.axis(axes)
  plt.xlabel("$x_1$", fontsize=18)
  plt.ylabel("$y$", rotation=0, fontsize=18)
  plt.legend(loc="upper center", fontsize=14)
```

```python
plt.figure(figsize=(5, 3))
plot_regression_predictions(tree_reg, x, y)
plt.show()
```
