```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('housing.csv')
data
```

```python
data.info() #data.isnull().any()
```

```python
data.dropna(inplace=True)
```

```python
data.info() #data.isnull().any()
```

```python
x = np.array(data)[:, :8].astype(np.float32)
x
```

```python
y = np.array(data)[:, 8].astype(np.float32)
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train[0], y_train[0]
```

```python
from sklearn.pipeline import Pipeline
```

# regressor

- Poly + Ridge
- SVR ( poly )
- SGD ( epsilon_insensitive )
- DecisionTreeRegressor

### Poly + Ridge

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge

ridge_reg = Pipeline([
  ('poly', PolynomialFeatures(degree=3, include_bias=False)),
  ('scal', StandardScaler()),
  ('ridge', Ridge(alpha=0.1, random_state=42))
])

ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_train, y_train), ridge_reg.score(x_test, y_test)
```

### SVR

```python
from sklearn.svm import SVR

poly_svr_reg = Pipeline([
  ('poly', PolynomialFeatures(degree=3, include_bias=False)),
  ('scal', StandardScaler()),
  ('svr', SVR(kernel="poly", degree=2, epsilon=0.1, gamma=1))
])

poly_svr_reg.fit(x_train, y_train)
```

```python
poly_svr_reg.score(x_train, y_train), poly_svr_reg.score(x_test, y_test)
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = Pipeline([
  ("poly", PolynomialFeatures(degree=2, include_bias=False)),
  ("scal", StandardScaler()),
  ("tree", DecisionTreeRegressor(max_depth=2, random_state=42))
])

tree_reg.fit(x, y)
```

```python
tree_reg.score(x_train, y_train)
```

### Ensemble Learning

```python
from sklearn.ensemble import VotingRegressor

voting_reg = VotingRegressor([
  ('ridge_reg', ridge_reg),
  ('poly_svr_reg', poly_svr_reg),
  # ('poly_sgd_reg', poly_sgd_reg),
  ('tree_reg', tree_reg)
])

voting_reg.fit(x_train, y_train)
```

```python
voting_reg.score(x_train, y_train), voting_reg.score(x_test, y_test)
```
