# Exploratory Data Analysis (EDA)
探索性資料分析，透過 `pandas`、`numpy`、`matplotlib`、`seaborn` 視覺化資料來幫助我們了解該資料集。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
print(type(iris))
```

```python
print(iris.DESCR)
```

```python
print("iris.data", iris.data[:5])
print("iris.target", iris.target)
print("iris.target_names", iris.target_names)
print("iris.feature_names", iris.feature_names)
```

```python
df_data = pd.DataFrame(data=np.c_[iris.data, iris.target] , columns=iris.feature_names + 
['target'])
df_data
```

```python
df_data.info()
```

```python
df_data.describe()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df_data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
plt.title('Box Plot of Iris Dataset Features')
plt.show()
```

```python
df_data.hist(alpha=0.5, layout=(5,5), figsize=(12, 12), bins=20)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 4)
fig.set_size_inches(15, 5)
sns.histplot(df_data, x='sepal length (cm)', hue='target', kde=True, ax=axes[0])
sns.histplot(df_data, x='sepal width (cm)', hue='target', kde=True, ax=axes[1])
sns.histplot(df_data, x='petal length (cm)', hue='target', kde=True, ax=axes[2])
sns.histplot(df_data, x='petal width (cm)', hue='target', kde=True, ax=axes[3])
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df_data, figsize=(10, 10), color='b', diagonal='kde')
```

```python
sns.pairplot(df_data, hue="target", height=2, diag_kind="kde")
```

```python
corr = df_data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','target']].corr()

corr
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, cbar=False)
```

```python
sns.lmplot(data=df_data, x='sepal length (cm)', y='sepal width (cm)', hue='target', fit_reg=False, legend=False)
plt.legend(title='target', labels=iris.target_names)
```

```python
sns.lmplot(data=df_data, x='petal length (cm)', y='petal width (cm)', hue='target', fit_reg=False, legend=False)
plt.legend(title='target', labels=iris.target_names)
```
